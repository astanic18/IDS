8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.




```python
#a)
mostCommonEthnicGroup = df["EthnicGroup"].mode()[0]
print(mostCommonEthnicGroup)
df["EthnicGroup"].fillna(mostCommonEthnicGroup, inplace=True)

mostCommonParentMaritalStatus = df["ParentMaritalStatus"].mode()[0]
print(mostCommonParentMaritalStatus)
df["ParentMaritalStatus"].fillna(mostCommonParentMaritalStatus, inplace=True)

mostCommonWeeklyStudyHours = df["WklyStudyHours"].mode()[0]
print(mostCommonWeeklyStudyHours)
df["WklyStudyHours"].fillna(mostCommonWeeklyStudyHours, inplace=True)

#b)
df["ParentEduc"].fillna("no formal education", inplace=True)

#c)
df["TestPrep"].fillna("none", inplace=True)

#d)
df.dropna(subset=["PracticeSport"], inplace=True)
df.dropna(subset=["IsFirstChild"], inplace=True)

df.reset_index(drop=True, inplace=True)

#e)
med = df["NrSiblings"].median()
df["NrSiblings"].fillna(med, inplace=True)

#f)
df["TransportMeans"].fillna("walking", inplace=True)

print(df.isna().sum())
```

9. Are there inconsistencies in the IsFirstChild column? Replace the inconsistent entries so they match the rest of the values. 

```python
print(df["IsFirstChild"])
print(df["IsFirstChild"].unique())

df["IsFirstChild"].replace({"nein": "no", 
                            "nope": "no", 
                            "si": "yes", 
                            "ye": "yes", 
                            "Yes": "yes", 
                            "yeah": "yes", 
                            "noo": "no", 
                            "ys": "yes", 
                            "yeah boy": "yes", 
                            "es": "yes"}, inplace=True)
print(df["IsFirstChild"].unique())
```

10. The final step to preparing the dataset for classification is transforming the non-numeric variables to numeric variables. 
For features Gender and LunchType use OneHot Encoder, and for EthnicGroup, ParentEduc, TestPrep, ParentMaritalStatus, PracticeSport, IsFirstChild, TransportMeans and WklyStudyHours use Label Encoder. 

```python
#transforming categorical data into numerical data
ohe = OneHotEncoder(sparse=False)
transformed = ohe.fit_transform(df[["Gender"]])
transformed = pd.DataFrame(transformed, columns=ohe.get_feature_names_out())
print(transformed)

df = pd.concat([df, transformed], axis=1)
print(df.columns)
df.drop("Gender", axis=1, inplace=True)
print(df.columns)
```

```python
ohe = OneHotEncoder(sparse=False)
transformed = ohe.fit_transform(df[["LunchType"]])
transformed = pd.DataFrame(transformed, columns=ohe.get_feature_names_out())
print(transformed)

df = pd.concat([df, transformed], axis=1)
print(df.columns)
df.drop("LunchType", axis=1, inplace=True)
print(df.columns)
```

```python
labelColumns = ["EthnicGroup", "ParentEduc", "TestPrep", "ParentMaritalStatus", "PracticeSport", "IsFirstChild", "TransportMeans", "WklyStudyHours"]
for column in labelColumns:
  le = LabelEncoder()
  transformed = le.fit_transform(df[column])
  transformed = pd.DataFrame(transformed, columns=[column])

  df[column] = transformed
```
